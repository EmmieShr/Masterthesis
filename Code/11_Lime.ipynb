```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
model_1_data = {
    'mutmaßlich': 7.1487e-02,
    'Mann': 5.0590e-02,
    'Tat': 4.7494e-02,
    'Der': 3.4266e-02,
    'psychischer': 3.3971e-02,
    'gefährden': 3.0644e-02,
    'der': 2.6002e-02,
    'polizeibekannt': 2.4154e-02,
    'aufgefallen': 2.3883e-02,
    'Menschen': 2.0146e-02
}

model_2_data = {
    'keine': -4.2085e-01,
    'mutmaßlich': 2.9933e-01,
    'legal': -1.5735e-01,
    'Obdachlosenunterkunft': -1.3201e-01,
    'afrikanische': -1.1943e-01,
    'genoss': -1.1115e-01,
    'war': -1.0926e-01,
    'gefährden': 1.0811e-01,
    'Behörden': 8.7179e-02,
    'Tat': 8.7175e-02
}

model_3_data = {
    'afrikanische': -5.0627e-08,
    'subsidiären': -3.9011e-08,
    'Zielgruppe': -2.5793e-08,
    'in': -2.0948e-08,
    'könnte': -2.0360e-08,
    'gefährden': -2.0319e-08,
    'registriert': -1.9825e-08,
    'psychischer': -1.9539e-08,
    'erstmals': -1.6560e-08,
    'Tat': 1.4735e-08
}

# Convert data into DataFrames
df1 = pd.DataFrame(list(model_1_data.items()), columns=['words', 'importance'])
df2 = pd.DataFrame(list(model_2_data.items()), columns=['words', 'importance'])
df3 = pd.DataFrame(list(model_3_data.items()), columns=['words', 'importance'])
```

```python
def assign_colors(importance):
    return ['green' if x > 0 else 'red' for x in importance]

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(20, 8))

# Plot for Model 1
sns.barplot(x='importance', y='words', data=df1, palette=assign_colors(df1['importance']), ax=axes[0])
axes[0].set_title("Explanation for BERT trained on real data annotated by humans")

# Plot for Model 2
sns.barplot(x='importance', y='words', data=df2, palette=assign_colors(df2['importance']), ax=axes[1])
axes[1].set_title("Explanation for BERT trained on synthetic data")

# Plot for Model 3
sns.barplot(x='importance', y='words', data=df3, palette=assign_colors(df3['importance']), ax=axes[2])
axes[2].set_title("Explanation for BERT trained on real data annotated by the LLM")

# Adjust layout
plt.tight_layout()
plt.show()
```
